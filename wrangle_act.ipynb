## Removing Retweets

Now, before the initial cleaning process, we need to remove retweets. Retweets can be identified by non-null values in the `retweeted_status_id`, `retweeted_status_user_id`, and `retweeted_status_timestamp` columns. Removing retweets is important because we only want to analyze original tweets to ensure the integrity and accuracy of our analysis. Retweets might duplicate the same tweet multiple times, skewing the results and insights derived from the data.

To remove the retweets, we'll filter out rows where these columns are not null, and then drop these columns since they will be empty after the removal. The following code achieves this:

```python
# Remove retweets
df_cleaned = df_cleaned[df_cleaned.retweeted_status_id.isnull()]
df_cleaned = df_cleaned[df_cleaned.retweeted_status_user_id.isnull()]
df_cleaned = df_cleaned[df_cleaned.retweeted_status_timestamp.isnull()]

# Drop retweet-related columns
df_cleaned = df_cleaned.drop(['retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp'], axis=1)

df_cleaned.info()
```

```python
from bs4 import BeautifulSoup

def extract_link_from_html(html_text):
    if pd.isnull(html_text):
        return None
    soup = BeautifulSoup(html_text, 'html.parser')
    link_tag = soup.find('a')
    if link_tag:
        return link_tag.get('href')
    else:
        return None

# Applying the function to extract links
df_cleaned['tweet_source'] = df_cleaned['tweet_source'].apply(extract_link_from_html)

```

### **Issue #2:**
**Merge Dog Breed Prediction Columns (p1, p1_conf, p1_dog, p2, p2_conf, p2_dog, p3, p3_conf, p3_dog)**

#### **Define:**

 Consolidate these columns into a more structured format to streamline the dataset and improve interpretability.


#### **Code:**
To clean the issue

```python
def get_top_prediction(row):
    predictions = [(row['first_prediction'], row['first_confidence']),
                   (row['second_prediction'], row['second_confidence']),
                   (row['third_prediction'], row['third_confidence'])]
    

    top_prediction = None
    top_confidence = -1  


    for prediction, confidence in predictions:
        if confidence > top_confidence:
            top_prediction = prediction
            top_confidence = confidence
    
    return top_prediction, top_confidence


df_cleaned['top_prediction'], df_cleaned['top_confidence'] = zip(*df_cleaned.apply(get_top_prediction, axis=1))


df_cleaned['is_dog_breed'] = ~df_cleaned['top_prediction'].isna()

columns_to_drop = ['first_prediction', 'first_confidence', 'is_dog_breed_1',
                   'second_prediction', 'second_confidence', 'is_dog_breed_2',
                   'third_prediction', 'third_confidence', 'is_dog_breed_3']
df_cleaned.drop(columns=columns_to_drop, inplace=True)
```

**4. Dog Ratings by Dog Stage:**
The ratings also vary by dog stage. Among the different stages, the "doggo" and "puppo" stages tend to have a higher median rating compared to other stages. This indicates that users might perceive these stages more favorably.

```python
dog_breed_counts = df_cleaned['top_prediction'].value_counts()

threshold = 10

common_breeds = dog_breed_counts[dog_breed_counts > threshold]

plt.figure(figsize=(12, 8))
sns.barplot(x=common_breeds.index, y=common_breeds.values, palette="viridis")
plt.title('Most Common Dog Breeds in Top Predictions')
plt.xlabel('Dog Breed')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
